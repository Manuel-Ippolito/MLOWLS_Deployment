# Exploration notebook 
Notebook for exploring the training data for the MLOWLS project.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
taxonomy = pd.read_csv("../data/taxonomy.csv")
print("Unique species in taxonomy.csv:", taxonomy["primary_label"].nunique())
```

```python
train = pd.read_csv("../data/birdclef-2025/train.csv")
print("Unique species in train_metadata.csv:", train["primary_label"].nunique())
```

```python
print("Label distribution in train_metadata.csv:")
print(train["common_name"].value_counts())
```

```python
plt.figure(figsize=(8, 40))
train["common_name"].value_counts().plot(kind="barh")
plt.title("Top 30 Species Distribution in Train Metadata")
plt.xlabel("Number of Occurrences")
plt.ylabel("Common Name")
plt.tight_layout()
plt.show()

print(f"Total species: {train['common_name'].nunique()}")
print(f"Most common species: {train['common_name'].value_counts().index[0]} ({train['common_name'].value_counts().iloc[0]} samples)")
print(f"Least common species: {train['common_name'].value_counts().index[-1]} ({train['common_name'].value_counts().iloc[-1]} samples)")
print(f"Mean samples per species: {train['common_name'].value_counts().mean():.1f}")
print(f"Median samples per species: {train['common_name'].value_counts().median():.1f}")
```
